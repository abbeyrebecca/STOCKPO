```python
ranking_df = pd.DataFrame({
    'Return': stocks_return,
    'Volatility': stocks_vol,
    'Sharpe Ratio': sharpe_ratio
})

# Sort by Sharpe Ratio descending (highest is best)
top_sharpe = ranking_df.sort_values(by='Sharpe Ratio', ascending=False)

# Display top 10 performers
print("🔝 Top 10 Stocks by Sharpe Ratio:")
print(top_sharpe.head(10))
```

## Data Integrity and Quality for Report

```python
import pandas as pd

dataagain = pd.read_csv('adjprice.csv')

#Load and check your dates
dataagain['Date'] = pd.to_datetime(dataagain['Date'])  # Ensure Date column is datetime

#Remove duplicates just in case, then set index
dataagain = dataagain.drop_duplicates(subset='Date')
dataagain.set_index('Date', inplace=True)

#Create a complete calendar date range (includes weekends)
full_calendar = pd.date_range(start=dataagain.index.min(), end=dataagain.index.max(), freq='D')

#make a list of all missing dates
missing_dates = full_calendar.difference(dataagain.index)

#count missing dates
if len(missing_dates) == 0:
    print("Your dataset includes all calendar days!")
else:
    missing_df = pd.DataFrame(index=missing_dates)
    missing_df['DayOfWeek'] = missing_df.index.day_name()
    missing_df['Type'] = missing_df['DayOfWeek'].apply(
        lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Likely Holiday'
    )

    # Print a summary
    print("Summary of missing dates:")
    print(missing_df['Type'].value_counts())
    print("\nSample of missing dates:")
    print(missing_df.head(10))
```
